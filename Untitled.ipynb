```python
import numpy as np
import random as rand
```

```python
def randint():
    return rand.randint(1, 100)
```

```python
a = np.array([randint(), randint()])
start = 2
S = 2*a.std()*a.std()
var = S/start
mu = a.mean()
print(var, a.std()*a.std())

for i in range(100):
    x_i = randint()
    a = np.append(a, x_i)
    mu_i = ((start+i)*mu + x_i)/(start+1+i)
    S = S + x_i*x_i - (start+1+i)*mu_i*mu_i + (start+i)*mu*mu
    var = S/(3+i)
    print(a.std()*a.std()- var)
    mu = mu_i
```
