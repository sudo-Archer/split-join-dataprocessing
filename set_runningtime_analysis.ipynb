```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def plot_with_mean(df):
    x = df.right_tree_size.unique()
    y = [df.loc[df['right_tree_size'] == xi, 'execution_time'].mean() for xi in x]
    plt.plot(x, y, 'x-')
    plt.ylim(bottom=0)
    plt.xlim(left=0)

def plot_with_median(df):
    x = df.right_tree_size.unique()
    y = [df.loc[df['right_tree_size'] == xi, 'execution_time'].median() for xi in x]
    plt.plot(x, y, '*-')
    plt.ylim(bottom=0)
    plt.xlim(left=0)
    
def plot_with_std(df):
    x = df.right_tree_size.unique()
    y = [df.loc[df['right_tree_size'] == xi, 'execution_time'].std() for xi in x]
    plt.plot(x, y, 'o-')
    plt.ylim(bottom=0)
    plt.xlim(left=0)
    
def plot_with_confidence(df):
    x = df.right_tree_size.unique()
    y = np.array([df.loc[df['right_tree_size'] == xi, 'execution_time'].mean() for xi in x])
    stdd = np.array([df.loc[df['right_tree_size'] == xi, 'execution_time'].std() for xi in x])
    lower = y-stdd
    upper = y+stdd
    plt.plot(x, y, '-')
    plt.ylim(bottom=0)
    plt.xlim(left=0)
    plt.fill_between(x, lower, upper, color='red', alpha=.5)
```
